```python
# 6. Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1.What is the class distribution (Phishing vs Legitimate)?
plt.figure(figsize=(6,4))
sns.countplot(x='CLASS_LABEL',data=data)
plt.title("Class Distribution")
plt.xlabel("CLASS_LABEL (0 = Legitimate, 1 = Phishing)")
plt.ylabel("Count")
plt.show()
```

Insight:
Phishing and legitimate websites are almost equally represented in this dataset. 
A balanced dataset helps machine learning models learn both behaviors fairly without 
bias toward one class.
    

```python
# 2.What is the distribution of URL Length?
plt.figure(figsize=(7,4))
plt.hist(data['UrlLength'], bins=30, edgecolor='black')
plt.title('URL Length Distribution')
plt.xlabel('UrlLength')
plt.ylabel('Frequency')
plt.show()
```

Insight:
Most URLs in the dataset have shorter lengths, showing that both phishing and legitimate sites usually avoid long URLs. A few URLs are extremely long, which may indicate phishing attempts where attackers add extra hidden components.

```python
#3. How does URL Length vary between classes?
plt.figure(figsize=(7,4))
sns.boxplot(x='CLASS_LABEL', y='UrlLength', data=data)
plt.title('UrlLength by Class')
plt.xlabel('CLASS_LABEL')
plt.ylabel('UrlLength')
plt.show()
```

Insight:
Phishing URLs are generally longer because attackers add extra paths, subdomains, or parameters to hide malicious intentions. 
Legitimate URLs stay shorter and more structured, making them easier for users to recognize.

```python
#4.Do phishing URLs contain more dots?
plt.figure(figsize=(6,4))
data.groupby('CLASS_LABEL')['NumDots'].mean().plot(kind='bar')
plt.title('Average NumDots by Class')
plt.xlabel('CLASS_LABEL')
plt.ylabel('NumDots')
plt.show()
```

Insight:
Phishing URLs usually contain more dots to make the structure look longer and confuse users about the true domain. 
Legitimate websites prefer shorter, clearer URLs that are easy to identify and trust.

```python
#5. How do special characters appear in both classes?
plt.figure(figsize=(8,4))
data.groupby('CLASS_LABEL')[['AtSymbol','NumUnderscore','NumDash']].mean().plot(kind='bar')
plt.title('Use of Special Characters by Class')
plt.xlabel('CLASS_LABEL')
plt.ylabel('Average Count')
plt.show()
```

Insight:
Phishing URLs tend to include more special characters like symbols and dashes to create misleading or complex links. 
Legitimate sites keep their URLs simple because recognizable structure builds trust for users.

```python
#6.Which features are highly correlated with CLASS_LABEL?
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(), cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

Insight:
Several URL-based features show a strong correlation with the phishing class, meaning they play a key role in distinguishing malicious links. These correlated attributes improve model learning by capturing consistent phishing patterns.

The heatmap helps identify which features are strongly linked with phishing behavior. Features that show stronger color intensity can be considered more important for training accurate ML models.

```python
#7. Which structural features have noticeable outliers?
plt.figure(figsize=(7,4))
sns.boxplot(data=data[['NumDash','NumDots','PathLevel']])
plt.title('Outliers in Structural Features')
plt.show()
```

Insight:
Some URLs show unusually high numbers of dashes, dots, and deeper path levels, indicating suspicious structural manipulation. 
These outliers are often crafted to confuse users and hide the true destination of phishing links.

```python
#8. Is HTTPS more common in legitimate URLs?
plt.figure(figsize=(6,4))
sns.countplot(data=data, x='NoHttps', hue='CLASS_LABEL')
plt.title('No HTTPS vs Phishing')
plt.xlabel('NoHttps (1 = No HTTPS, 0 = HTTPS)')
plt.ylabel('Count')
plt.show()
```

Insight:
Websites without HTTPS are more frequently associated with phishing since attackers skip SSL certificates to reduce cost and effort.
Legitimate websites typically use HTTPS to ensure security and user trust during communication.

```python
#9. Do phishing websites use more query components?
plt.figure(figsize=(7,4))
sns.boxplot(x='CLASS_LABEL', y='NumQueryComponents', data=data)
plt.title('Query Components vs Class')
plt.xlabel('CLASS_LABEL')
plt.ylabel('NumQueryComponents')
plt.show()
```

Insight:
Phishing URLs often include more query components to encode redirects or hide malicious parameters. 
Legitimate websites keep query structures minimal and straightforward for clarity and usability.

```python
#10. Compare distribution of number of subdomains in classes
plt.figure(figsize=(7,4))
sns.stripplot(x='CLASS_LABEL', y='SubdomainLevel', data=data, jitter=True)
plt.title('Subdomain Usage by Class')
plt.xlabel('CLASS_LABEL')
plt.ylabel('SubdomainLevel')
plt.show()
```

Insight:
Phishing URLs tend to contain a higher number of subdomains, creating long and confusing structures to mislead users.
Legitimate websites usually maintain simple and recognizable domain formats.

```python
#11. Do phishing pages disable right-click more often?
plt.figure(figsize=(6,4))
sns.countplot(x='RightClickDisabled', hue='CLASS_LABEL', data=data)
plt.title('Right-Click Disabled vs Class')
plt.xlabel('RightClickDisabled (1 = Disabled)')
plt.ylabel('Count')
plt.show()
```

```python
#11. Do phishing pages disable right-click more often?
plt.figure(figsize=(6,4))
sns.countplot(x='RightClickDisabled', hue='CLASS_LABEL', data=data)
plt.title('Right-Click Disabled vs Class')
plt.xlabel('RightClickDisabled (1 = Disabled)')
plt.ylabel('Count')
plt.show()
```

Insight:
Phishing pages often disable right-click to prevent users from inspecting elements or revealing suspicious code. 
Legitimate websites allow normal browsing controls because they have nothing to hide.

```python
#12. Are popup windows more common in phishing attempts?
plt.figure(figsize=(6,4))
sns.countplot(x='PopUpWindow', hue='CLASS_LABEL', data=data)
plt.title('Pop-Up Window Usage vs Class')
plt.xlabel('PopUpWindow (1 = Popup Appears)')
plt.ylabel('Count')
plt.show()
```

Insight:Phishing websites tend to use pop-ups more often to trigger fake alerts or force users into quick actions without thinking. 
Legitimate sites rarely rely on pop-ups since they disrupt users and reduce trust.
